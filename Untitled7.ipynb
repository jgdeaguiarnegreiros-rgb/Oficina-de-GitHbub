```python
from abc import ABC, abstractmethod

# --- ABSTRAÇÕES (Interfaces) ---

class FreteStrategy(ABC):
    @abstractmethod
    def calcular(self) -> float: pass

class DescontoStrategy(ABC):
    @abstractmethod
    def calcular(self, total: float) -> float: pass

class MetodoPagamento(ABC):
    @abstractmethod
    def pagar(self, valor: float): pass

# --- IMPLEMENTAÇÕES ESPECÍFICAS ---

class FreteSedex(FreteStrategy):
    def calcular(self): return 25.0

class DescontoVIP(DescontoStrategy):
    def calcular(self, total): return total * 0.15

class PagamentoPix(MetodoPagamento):
    def pagar(self, valor): print(f"Pagamento de R${valor:.2f} via PIX realizado.")

class ServicoEmail:
    def enviar(self): print("E-mail de confirmação enviado.")

class ServicoFiscal:
    def gerar_nota(self): print("Nota fiscal gerada com sucesso.")

# --- SISTEMA PRINCIPAL ---

class LojaOnline:
    def __init__(self):
        self.produtos = []
        self.total = 0
        self.email = ServicoEmail()
        self.fiscal = ServicoFiscal()

    def adicionar_produto(self, nome, preco):
        self.produtos.append({"nome": nome, "preco": preco})
        self.total += preco

    def finalizar_compra(self, frete: FreteStrategy, desconto: DescontoStrategy, pagamento: MetodoPagamento):
        valor_desconto = desconto.calcular(self.total)
        valor_frete = frete.calcular()
        valor_final = self.total - valor_desconto + valor_frete

        print(f"Subtotal: {self.total} | Final: {valor_final}")

        pagamento.pagar(valor_final)
        self.fiscal.gerar_nota()
        self.email.enviar()
```
