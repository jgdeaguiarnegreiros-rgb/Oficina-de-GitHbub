Explicação da Aplicação dos Princípios
SRP (Responsabilidade Única): A lógica de frete, desconto, pagamento e notificação foi movida para classes próprias. A LojaOnline agora apenas **coordena** o fluxo da venda.

OCP (Aberto/Fechado): Para adicionar um novo tipo de frete (ex: "Entrega por Drone"), criamos uma nova classe que herda de FreteStrategy. Não mexemos em uma única linha da classe LojaOnline.

LSP (Substituição de Liskov): As subclasses de frete e desconto podem ser trocadas entre si sem quebrar o funcionamento do método finalizar_compra.

ISP (Segregação de Interface): Criamos interfaces específicas para cada comportamento (Frete, Pagamento, Desconto), evitando que uma classe seja obrigada a implementar métodos que não utiliza.

DIP (Inversão de Dependência): A LojaOnline não depende mais de "strings" ou implementações concretas, mas sim das abstrações (FreteStrategy, etc.). As dependências são injetadas no método.
